```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
rating=pd.read_csv('rating.csv')
plt.figure(figsize=(9,5))
k=rating.sort_values(by="Price_range _rating")

k.plot.bar(y="Count of costly restaurants",x="Price_range _rating",width=0.6,color='blue',alpha=0.7);
plt.ylabel("Count of costly Restaurants",fontsize=13)
plt.xlabel("Price Range Rating",fontsize=13)
plt.xticks(fontsize=10)

plt.xticks(fontsize=10,rotation=360)
```

```python
rest=pd.read_csv('rest.csv')
```

```python
sns.light_palette("navy", reverse=True)

plt.figure(figsize=(9,5))
k=rest.sort_values(by="Num_of_Restaurants")

sns.barplot(x="city",y="Num_of_Restaurants",data=k,palette='Set1');
plt.xlabel("City",fontsize=13)
plt.ylabel("Number of Restaurants",fontsize=13)
plt.xticks(fontsize=10,rotation=45)

```
